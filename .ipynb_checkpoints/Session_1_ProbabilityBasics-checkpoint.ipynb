# Probability Basics

***Key Terms***


*   **Random Experiment**
Mechanism or steps to produce an outcome. 

eg.  Tossing of a Coin, Rolling of a dice, Picking a food joint for audit in Chandigarh, Selecting 100 employees from a company for survey

*   **Outcome**
Result of Random Experiment

eg. [Occurence of Head]

 [Dice landing with number 4 face]
 
 [McDonald's at Chandigarh Sector 16 for Audit]
 
 [6 Employees from Project A, 2 employees from Project B, 4 employees from Project C,... 2 employees from Project Z]

*   **Sample Space**
Set of all possible outcomes

eg. {Heads,Tails}

{1,2,3,4,5,6}

{All food joints in city of Chandigarh}

{All possible combinations of picking up 100 employees from a company}

*   **Event**
Subset of sample space



image.png

Random Experiment: Rolling of two dice

Outcome: for eg 5,5

Sample Space: {[1,1], [1,2], ....[3,4],[3,5],[3,6], ..., [6,4],[6,5],[6,6]}]

Event: for eg. where dice 1 has 2 on the face

***Remember***
For any event E

***0  <= P(E) <=  1***

**Cumulative Frequency Distribution**

The **cumulative frequency (c.f)** is the total of frequencies, where the frequency of the first data point/class interval is added to the frequency of the second data point/class interval and then the sum is added to the frequency of the third data point/class interval and so on.

A table that shows the cumulative frequencies, which are distributed over different classes is known as the **cumulative frequency table** or **cumulative frequency distribution**.

# Random Variables

**WHAT?**

Random Variable (R.V.) is a real valued variable whose numerical value is determined by an underlying experiment.

eg.


1.   Number of Tails in 1000 coin tosses
2.   Number of wide balls in a cricket match
3.   Number of goals scored in a football match



---
A random variable X is a function from the sample space to the real numbers.
> > > > X:S→R





image.png


Sample space is a collection of all possible results of an experiment.
Eg. 

In tossing a coin we could get a heads or a tails {H,T}

If we toss two coins possible results would be {HH, HT, TH, TT}  we can denote them as 0,1,2,3  (some numerical value)

Now Let X be the RV which measures the outcome of tossing of two coins.

Experiment:  Tossing of two coins
Sample Space: {HH, HT, TH, TT}
X: RV which measures the outcome

Toss1: HH
Toss2: HH
Toss3: TH
Toss4: TT

So HH, HH, TH, TT are outcomes measured by R.V. X

Range of X is {0,1,2,3}

and X takes values 0,0,2,3 respectively

**WHY?**

As a data scientist you are interested to know which values are more likely or less likely to occur so that you can take informed decisions.


You may find applications in 


1.   Exploratory Data Analysis
2.   Formulating Hypothesis to understand if there is a relationship between two features

**HOW?**

**Discrete Random Variable**

A R.V. X that can take up only a countable number of values is called a Discrete Random Variable.


eg.


*   For a coin toss the outcomes could be {H, T} which could be mapped to numbers {0,1}.

So R.V. say X can take values 0 or 1
*   For roll of a dice the outcomes could be {1,2,3,4,5,6} and R.V. Y could take any of these 6 values

* For Age (in ordinal) could take values from {Baby, Chid, Teen, Adult} so R.V. A could be mapped to {0,1,2,3} and A could take any of these values


***Any variable X is a discrete random variable that maps outcome(s)of an experiment to a countable set of numbers. We call this set Rx or Range of X***

Rx = {x1, x2,...,xn}



**Probability Mass Function**

But what to do with X and Rx

Our interest is in understanding that what is the chance(probability) that X will take a value x1, or x2, or x3, or  ,..., or xn.

For instance,

in LungCap data we have AgeGrp as {Baby/Toddler, Child, Teen, Young Adult}

If we know that P(X=Teen) smoking is high, say 0.67 then we can take corrective actions to inform / guide them or send them to rehab.


***In effect here you are trying to explore the relationship between two variables.***

Does age have anything to do with smoking? Does smoking have anything to do with LungCap?




For Discrete RV

1.   represents the probabilities of the different outcomes for a discrete random variable
2.   it allows us to understand the pattern of the different outcomes in the random variable
3. the representation of all the probabilities of a random variable put together

***Properties of PMF***

Described as p(x) or P(X=x)


1.   0 <=  P(X=x) <= 1 for all x in Rx 
2.   P(x1) + P(x2) + P(x3) +...+ P(xn) =1  for all x in Rx 
*i.e. some of probabilities of all possible values of X will always be 1*
3.  For a subset A of Rx  : A = {xi, xi+1, ...xj) 
P(X in A) = P(xi) + P(xi+1) +...+ P(xj)

**Continuous Random Variable**

A R.V. X that can take up infintely many values within a range is called a Continuous Random Variable.
Its distribution is denoted by PDF or Probability Density Function

eg.

*   Height of students of Std 5 may take any value in range Rx{130,145} cms.
So R.V. say X can take values in Rx

* For LungCap (in continuous) could take values from Rl: [0.5-15] so R.V. L could be mapped to any real number in Rl and A could take any of these values.


*Please  note that these variables could take any real values between -inf to inf but you will observe that in real life most values will lie within a range and maybe domain specific. For instance Age can take values from [0-110] or height may take values from [40-220]*


***Any variable X is a continous random variable that maps outcome(s)of an experiment to any real number within range Rx of X.***

Rx = {x1, x2,...,xn}

**Empirical Distribution**

An empirical distribution is one for which each possible event is assigned a
probability derived from experimental observation. It is assumed that the events
are independent and the sum of the probabilities is 1.

Thus, PMF of a Discrete R.V. (discrete description) or PDF of a Continous R.V. (continuous description) are both empirical distributions.
